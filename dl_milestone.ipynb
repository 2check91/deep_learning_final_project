```python
import numpy as np
from PIL import Image
import os,sys
from skimage import data, io, filters, transform
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import DBSCAN
from sklearn.neighbors import LSHForest
from sklearn.neural_network import MLPClassifier
```

```python
raw_hoods = io.imread_collection('/home/ubuntu/Github/deep_learning_final_project/pictures/hoods/full/*')
unraw_hoods = [transform.resize(i, (200,200)) for i in raw_hoods]
y_hoods = [0]*150
```

```python
raw_fenders = io.imread_collection('/home/ubuntu/Github/deep_learning_final_project/pictures/fenders/full/*')
unraw_fenders = [transform.resize(i, (200,200)) for i in raw_fenders]
y_fenders = [1]*150
```

```python
raw_bumpers = io.imread_collection('/home/ubuntu/Github/deep_learning_final_project/pictures/bumpers/full/*')
unraw_bumpers = [transform.resize(i, (200,200)) for i in raw_bumpers]
y_bumpers = [2]*150
```

```python
Y = np.array(y_hoods + y_fenders + y_bumpers)
X = np.array(unraw_hoods + unraw_fenders + unraw_bumpers)
```

```python
X = np.reshape(X,[450,200*200*3])
#X = filters.sobel(X)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
clf = SVC(kernel='poly', degree=4)
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test)
```

```python
np.mean(y_preds == y_test)
```

```python
rcl = RandomForestClassifier(n_estimators=100, max_depth=120000, criterion='entropy', min_samples_split=2)
rcl.fit(X_train,y_train)
```

```python
tree_preds = rcl.predict(X_test)
```

```python
np.mean(tree_preds == y_test)
```
